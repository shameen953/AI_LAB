```python
#Home Assignment (Sameed - 128)
import pandas as pd
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split

file_path = 'AILab3_data.xlsx'
df = pd.read_excel(file_path)

# Display the original dataset
print("Original Dataset:")
print(df)

# Label encoding for categorical columns
le_gender = preprocessing.LabelEncoder()
le_alcohol = preprocessing.LabelEncoder()
le_smoking = preprocessing.LabelEncoder()
le_sleep_disorder = preprocessing.LabelEncoder()

# Encode categorical features
df['Gender'] = le_gender.fit_transform(df['Gender'])
df['Alcohol Consumption'] = le_alcohol.fit_transform(df['Alcohol Consumption'])
df['Smoking Habit'] = le_smoking.fit_transform(df['Smoking Habit'])
df['Sleep Disorder'] = le_sleep_disorder.fit_transform(df['Sleep Disorder'])

# Display the encoded dataset
print("\nEncoded Dataset:")
print(df)

# Assign features and label variables
features = df[['Gender', 'Age', 'Physical Activity Level', 'Stress Level', 'Alcohol Consumption', 'Smoking Habit', 'Work Hours']]
labels = df['Sleep Disorder']

# Split dataset into training and testing sets
features_train, features_test, label_train, label_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Generate a model using K-Neighbors classifier
model = KNeighborsClassifier(n_neighbors=3, metric='euclidean')

# Fit the dataset on classifier
model.fit(features_train, label_train)

# Perform prediction
predicted = model.predict(features_test)

# Print predictions
print("\nPredictions:", predicted)

# Confusion Matrix with labels
labels = [0, 1]  # Assuming binary classification with labels 0 and 1
conf_mat = confusion_matrix(label_test, predicted, labels=labels)
print("\nConfusion Matrix:")
print(conf_mat)

# Accuracy
accuracy = accuracy_score(label_test, predicted)
print("Accuracy:", accuracy)

# Interpretation of Confusion Matrix
if conf_mat.shape[0] > 1 and conf_mat.shape[1] > 1:
    TP = conf_mat[1][1] if conf_mat.shape[0] > 1 else 0  # True Positives
    FP = conf_mat[0][1] if conf_mat.shape[0] > 1 else 0  # False Positives
    TN = conf_mat[0][0] if conf_mat.shape[0] > 0 else 0  # True Negatives
    FN = conf_mat[1][0] if conf_mat.shape[0] > 1 else 0  # False Negatives

    print("\nConfusion Matrix Interpretation:")
    print(f"True Positives (TP): {TP}")
    print(f"False Positives (FP): {FP}")
    print(f"True Negatives (TN): {TN}")
    print(f"False Negatives (FN): {FN}")

    # Accuracy Calculation
    print("\nAccuracy Calculation:")
    print(f"Accuracy = (TP + TN) / (TP + TN + FP + FN) = ({TP} + {TN}) / ({TP} + {TN} + {FP} + {FN})")
    accuracy_manual = (TP + TN) / (TP + TN + FP + FN)
    print(f"Calculated Accuracy: {accuracy_manual:.2f} or {accuracy_manual * 100:.2f}%")
else:
    print("Confusion matrix does not have enough classes to compute TP, FP, TN.")
```

```python
import pandas as pd
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split

# Load the dataset
file_path = 'LAB3_data.xlsx'
df = pd.read_excel(file_path)

# Display the dataset
print("Original Dataset:")
print(df)

# Label encoding for categorical columns
le_gender = preprocessing.LabelEncoder()
le_sleep_disorder = preprocessing.LabelEncoder()

df['Gender'] = le_gender.fit_transform(df['Gender'])
df['Sleep Disorder'] = le_sleep_disorder.fit_transform(df['Sleep Disorder'])

# Display the encoded dataset
print("\nEncoded Dataset:")
print(df)

# Assign features and label variables
features = df[['Gender', 'Age', 'Physical Activity Level', 'Stress Level']]
labels = df['Sleep Disorder']

# Split dataset into training and testing sets
features_train, features_test, label_train, label_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Generate a model using K-Neighbors classifier
model = KNeighborsClassifier(n_neighbors=3, metric='euclidean')

# Fit the dataset on classifier
model.fit(features_train, label_train)

# Perform prediction
predicted = model.predict(features_test)

# Print predictions
print("\nPredictions:", predicted)

# Ensure both label_test and predicted are of the same type
label_test_encoded = label_test.to_numpy()  # Convert to numpy array if needed

# Convert the predicted values to the same format as label_test
predicted_encoded = predicted.astype(int)

# Define all possible labels
all_labels = le_sleep_disorder.classes_

# Confusion Matrix with labels
conf_mat = confusion_matrix(label_test_encoded, predicted_encoded, labels=all_labels)
print("\nConfusion Matrix:")
print(conf_mat)

# Accuracy
accuracy = accuracy_score(label_test_encoded, predicted_encoded)
print("Accuracy:", accuracy)

# Interpretation of Confusion Matrix
if conf_mat.shape[0] > 1 and conf_mat.shape[1] > 1:
    TP = conf_mat[1][1]  # True Positives
    FP = conf_mat[0][1]  # False Positives
    TN = conf_mat[0][0]  # True Negatives
    FN = conf_mat[1][0]  # False Negatives

    print("\nConfusion Matrix Interpretation:")
    print(f"True Positives (TP): {TP}")
    print(f"False Positives (FP): {FP}")
    print(f"True Negatives (TN): {TN}")
    print(f"False Negatives (FN): {FN}")

    # Accuracy Calculation
    print("\nAccuracy Calculation:")
    print(f"Accuracy = (TP + TN) / (TP + TN + FP + FN) = ({TP} + {TN}) / ({TP} + {TN} + {FP} + {FN})")
    accuracy_manual = (TP + TN) / (TP + TN + FP + FN)
    print(f"Calculated Accuracy: {accuracy_manual:.2f} or {accuracy_manual * 100:.2f}%")
else:
    print("\nWarning: Confusion matrix does not have enough classes to compute TP, FP, TN.")
```
